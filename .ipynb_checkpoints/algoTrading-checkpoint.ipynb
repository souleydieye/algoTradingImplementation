```python
# Here we implement a function that gives the maximum return given values for sma and lma windows
symbol = "MSFT.O"

def get_max_return(s_window, l_window, dataframe) :
    
    sma_ct_vec = dataframe.rolling(window = s_window, center = False).mean()
    sma_lt_vec = dataframe.rolling(window = l_window, center = False).mean()
    
    # position vector
    positions_ = sma_ct_vec > sma_lt_vec
    positions_ = 2*positions_ - 1 #mapping from 0-1 to -1-1

    
    # computing profit
    log_vals_ = np.log(dataframe[symbol])
    returns_vec_ = log_vals_.diff()
    profits_ = np.multiply(positions_[symbol], returns_vec_)
    
    expected_prof = profits_.sum()
    
    print ('Profit for s_window %d and l_window %d is : %.4f'%(s_window, l_window, expected_prof))
    
    return expected_prof
    
    
```

```python
#tr_eikon_path = '/Users/adnanebazouz/Desktop/algo_trading/EURUSD1440.csv'

##tr_eikon_path = '/Users/adnanebazouz/Desktop/algo_trading/tr_eikon_eod_data.csv'
df_tr_eikon = pd.read_csv(tr_eikon_path, index_col = 0, parse_dates = True).dropna()

#symbol = "AAPL.O"
aappl_df = pd.DataFrame(df_tr_eikon[symbol]).dropna()


sma_window = np.arange(10, 81, 4)
lma_window = np.arange(120, 300, 10)

# for each (p, q) we get the maximum return and save it in return matrix
return_matrix = np.zeros((len(sma_window), len(lma_window)))

for p, s_wind in enumerate(sma_window) : 

    for q, l_wind in enumerate(lma_window) : 
        
        return_matrix[p, q] = get_max_return(int(s_wind), int(l_wind), dataframe = aappl_df )


smax_position, lmax_position = np.where(return_matrix == return_matrix.max())
print ("Max return is for sma_window %d and lma_window %d : %.4f"%(sma_window[smax_position], 
                                                                   lma_window[lmax_position],
                                                                  return_matrix.max()))
```

# Fibonacci Trading Strategies

The Fibonacci ratios, 23.6%, 38.2%, and 61.8%, can be applied for time series analysis to find support level. Whenever the price moves substantially upwards or downwards, it usually tends to retrace back before it continues to move in the original direction. For example, if the stock price has moved from \\$200 to \\$250, then it is likely to retrace back to \\$230 before it continues to move upward. The retracement level of \\$230 is forecasted using the Fibonacci ratios.

We can arrive at \\$230 by using a simple math

Total up move = \\$250 – \\$200 = \\$50

38.2% of up move = 38.2% * 50 = \\$19.1

Retracement forecast = \\$250 – \\$19.1 = \\$230.9

Any price level below \\$230 provides a good opportunity for the traders to enter into new positions in the direction of the trend. Likewise, we can calculate for 23.6%, 61.8% and the other Fibonacci ratios.

#### How to find Fibonacci retracement levels?


As we now know, retracements are the price movements that go against the original trend. To forecast the Fibonacci retracement level we should first identify the total up move or total down move. To mark the move, we need to pick the most recent high and low on the chart.

```python
def fibonacciRetracementStrategy(ratio, symbol, data) :
    dataframe = pd.DataFrame(data[symbol]).dropna()
    dataframe['pastWave'] = dataframe.shift(1) - dataframe.shift(2)
    dataframe['potentialRetracementLevel'] = dataframe[symbol].shift(1) - ratio*dataframe['pastWave']
    #aappl_df['retracementLevel'] = np.where(aappl_df['wave']>=0, aappl_df['AAPL.O']-o.382* aappl_df['wave'], aappl_df['AAPL.O']-3* aappl_df['wave'])
    dataframe['variation'] = dataframe[symbol].diff()
    dataframe.loc[((dataframe['pastWave'] > 0) & ((dataframe['variation'] < 0) & (dataframe[symbol] > dataframe['potentialRetracementLevel']))), 'position'] = -1
    dataframe.loc[((dataframe['pastWave'] < 0) & ((dataframe['variation'] > 0) & (dataframe[symbol] < dataframe['potentialRetracementLevel']))), 'position'] = 1
    dataframe.loc[(((dataframe['pastWave'] > 0) & (dataframe['variation'] < 0) & (dataframe[symbol] < dataframe['potentialRetracementLevel'])) | ((dataframe['pastWave'] > 0) & (dataframe['variation'] > 0))), 'position'] = 1
    dataframe.loc[(((dataframe['pastWave'] < 0) & (dataframe['variation'] > 0) & (dataframe[symbol] > dataframe['potentialRetracementLevel'])) | ((dataframe['pastWave'] < 0) & (dataframe['variation'] < 0))), 'position'] = -1
    log_vals_ = np.log(dataframe[symbol])
    returns_vec_ = log_vals_.diff()
    dataframe['profit'] = np.multiply(dataframe['position'], returns_vec_)

    return dataframe, dataframe['profit'].sum()
```

```python
def seeOrders(ratio, symbol, df):
    plt.figure(figsize=(15,7))
    dataframe = fibonacciRetracementStrategy(ratio, symbol, df)[0]
    print("the profit is:  ", fibonacciRetracementStrategy(ratio, symbol, df)[1])
    dataframe[symbol].plot()
    buy = dataframe[symbol].copy()
    sell = dataframe[symbol].copy()
    buy[(((dataframe['pastWave'] < 0) & ((dataframe['variation'] > 0) & (dataframe[symbol] < dataframe['potentialRetracementLevel']))) | (((dataframe['pastWave'] > 0) & (dataframe['variation'] < 0) & (dataframe[symbol] < dataframe['potentialRetracementLevel'])) | ((dataframe['pastWave'] > 0) & (dataframe['variation'] > 0))))] = 0
    sell[(((dataframe['pastWave'] > 0) & ((dataframe['variation'] < 0) & (dataframe[symbol] > dataframe['potentialRetracementLevel']))) | (((dataframe['pastWave'] < 0) & (dataframe['variation'] > 0) & (dataframe[symbol] > dataframe['potentialRetracementLevel'])) | ((dataframe['pastWave'] < 0) & (dataframe['variation'] < 0))))] = 0
    buy.plot(color='g', linestyle='None', marker='x')
    sell.plot(color='r', linestyle='None', marker='.')
    x1,x2,y1,y2 = plt.axis()
    plt.axis((x1,x2,dataframe[symbol].min(),dataframe[symbol].max()))
    plt.legend([symbol, "Buy Signal", 'Sell Signal'])
    plt.show()
```

# Pairs Trading Strategy

Definition:
<p style="text-align:center">
 Two series $ X_t = (X_1(t),...,X_n(t))~ and ~ Y_t = (Y_1(t),...,Y_n(t)) $ are cointegrated if the ratio $ Y/X $ vary around a mean $\alpha$  </p>  
 
__**Principle of Pairs trading:**__

First of all, we have to find a cointegrated pairs.
As our cointegrated pairs drift towards and apart from each other then we have a variation of the of the spread.
In order to make a pairs trade, we have to buy one of the securities and sell the other one.
This Strategy is a market neutral strategy, it allows to make profit from any market trend.
Because:
> * If both seurities move on the same direction, we neither gain or lose money
> * If $ Y/X > \alpha $ , then we expect a decrease of the ratio $ Y/X $ . So we will sell $X$ and buy $Y$
> * If $ Y/X < \alpha $ , then we expect an increase of the ratio $ Y/X $ . So we will buy $X$ and sell $Y$ 

## CoIntegrated Pairs research

In this part we are going to compute a statistical test between all the Pairs of our dataset in order to find the co-integrated pairs. 
We will define a threshold of 1% and we will consider only the pairs for which the P-value is under the threshold.

```python
def coIntegratedPairsResearch(dataframe):
    n = dataframe.shape[1]
    pValueMatrix = np.ones((n, n)) 
    indicesList = list(dataframe)
    coIntegratedPairsList = [] # initilize the list for cointegration
    for i in range(n):
        for j in range(i+1, n): # for j bigger than i
            result = coint(dataframe[indicesList[i]], dataframe[indicesList[j]]) # get conintegration
            pvalue = result[1] # get the pvalue
            pValueMatrix[i, j] = pvalue
            if pvalue < 0.01: 
                coIntegratedPairsList.append((indicesList[i], indicesList[j], pvalue)) # record the contract with that p-value
    coIntegratedPairsList.sort(key=lambda x:x[2])
    return pValueMatrix, coIntegratedPairsList
```

```python
def correlationPairsResearch(dataframe):
    n = dataframe.shape[1]
    correlationMatrix = np.ones((n, n)) 
    indicesList = list(dataframe)
    correlationPairsList = [] # initilize the list for cointegration
    for i in range(n):
        for j in range(i+1, n): # for j bigger than i
            pvalue = stats.pearsonr(dataframe[indicesList[i]], dataframe[indicesList[j]])[0] # get conintegration
            correlationMatrix[i, j] = pvalue
            if abs(pvalue) > 0.9: 
                correlationPairsList.append((indicesList[i], indicesList[j], pvalue)) # record the contract with that p-value
    correlationPairsList.sort(key=lambda x:-x[2])
    return correlationMatrix, correlationPairsList
```

```python
def pairsTradingStrategy(data, symbols, window, bollingerBand, exitScore = 0):
    
    model = sm.OLS(data[symbols[0]], data[symbols[1]])
    result = model.fit()
    dataframe = pd.DataFrame({'stock1':data[symbols[0]],'stock2':data[symbols[1]], 'spread': data[symbols[0]] - (data[symbols[0]] * result.params[0])})
    meanSpread = dataframe.spread.rolling(window=window).mean()
    stdSpread = dataframe.spread.rolling(window=window).std()    
    dataframe['zScore'] = (dataframe.spread-meanSpread)/stdSpread 
    dataframe['sameTrends'] = np.sign(dataframe['stock1'].shift(1) - dataframe['stock1'])*np.sign(dataframe['stock2'].shift(1) - dataframe['stock2'])
    dataframe['orderStock1'] = 0
    dataframe['orderStock2'] = 0


    dataframe.loc[((dataframe['zScore'] >  bollingerBand) & (dataframe['zScore'].shift(1) < bollingerBand) & (dataframe['sameTrends'] == -1)), 'orderStock1'] = -1
    dataframe.loc[((dataframe['zScore'] < - bollingerBand) & (dataframe['sameTrends'] == -1)), 'orderStock1'] = 1
    dataframe.loc[((dataframe['zScore'] < - bollingerBand) & (dataframe['zScore'].shift(1) > - bollingerBand) & (dataframe['sameTrends'] == -1)), 'orderStock2'] = 1
    dataframe.loc[((dataframe['zScore'] < - bollingerBand) & (dataframe['sameTrends'] == -1)), 'orderStock2'] = -1
     

        
    dataframe['stock1'] = np.log(dataframe['stock1'])
    dataframe['stock1'] = dataframe['stock1'].diff()
    dataframe['stock2'] = np.log(dataframe['stock2'])
    dataframe['stock2'] = dataframe['stock2'].diff()   
    dataframe['profit'] = np.multiply(dataframe['stock1'], dataframe['orderStock1']) + np.multiply(dataframe['stock2'], dataframe['orderStock2'])
    return dataframe, dataframe['profit'].sum()
```
